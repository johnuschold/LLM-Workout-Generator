```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_huggingface.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
import pandas as pd
import os # Import os for path manipulation

```

```python
# --- Exercise Data Processing ---
exercises = pd.read_csv("data/exercises_data_final.csv")
encoding='utf-8-sig'

# Save the 'description' column to a text file.
exercises["description"].to_csv("exercise_descriptions.txt",
                                   index=False,
                                   header=False,
                                   lineterminator='\\n')

```

```python
# Load the descriptions from the text file.
raw_documents_exercises = TextLoader("data/exercise_descriptions.txt", encoding='utf-8').load()

# Split exercise documents.
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\\n")
documents_exercises = text_splitter.split_documents(raw_documents_exercises)


```

```python
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# Create FAISS vector store for exercises
vectorstore_exercises = FAISS.from_documents(documents_exercises, embeddings)

# Save exercise vector store
save_path_exercises = "./faiss_index_exercises"
vectorstore_exercises.save_local(save_path_exercises)
print(f"Exercise vector store saved to: {os.path.abspath(save_path_exercises)}")

```

```python
query = "jumping exercise"
docs = vectorstore_exercises.similarity_search(query, k=1)
docs
```
